```python
#Gender
```

```python
palette = sns.color_palette("inferno", n_colors=2)
ax = sns.countplot(x= 'Gender', data = df, palette = palette)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sales_gen
```

```python
palette = sns.color_palette("inferno", n_colors=2)
sns.set({'figure.figsize': (7, 5)})
sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen, palette =palette)
```

```python
#From the above graphs we can see that most of the buyers are females & even te purchasing power of famels are greater then male
```

```python
#Age
```

```python
palette = sns.color_palette("inferno", n_colors=2)
ax = sns.countplot(data= df, x = 'Age Group' , hue = 'Gender', palette = palette)

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sales_state = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values( by = 'Amount', ascending=False)
palette = sns.color_palette("inferno", n_colors=7)
sns.barplot(x = 'Age Group', y = 'Amount', data = sales_state, palette = palette)
```

```python
#from above graphs we can see that most of the buyers are of the ga egroup between 26-35 years female
```

```python
#State
```

```python
df.columns
```

```python
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
sns.set(rc = {'figure.figsize':(16,8)})
palette = sns.color_palette("inferno", n_colors=10)
sns.barplot( x= 'State', y= 'Orders', data = sales_state, palette= palette)
```

```python
sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by= 'Amount', ascending = False).head(10)

sns.set({'figure.figsize': (16, 8)})
palette = sns.color_palette("inferno", n_colors=10)
sns.barplot(x= 'State', y='Amount', data= sales_state, palette = palette)
```

```python
#from above graphs we can see that most of the orders are from Uttar Pradesh, Maharashtra & Karnataka respectively
```

```python
#Martital Status
```

```python
palette = sns.color_palette("inferno", n_colors=2)
sns.set({'figure.figsize': (7, 5 )})
ax = sns.countplot (x = 'Marital_Status', data = df, palette = palette)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#occupation
```

```python
sns.set({'figure.figsize': (20, 5)})
palette = sns.color_palette("inferno", n_colors=15)
ax = sns.countplot(x = 'Occupation', data = df, palette= palette)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occupation = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False)
palette = sns.color_palette("inferno", n_colors=15)
sns.set({'figure.figsize': (20, 5)})
sns.barplot(x = 'Occupation', y='Amount', data = sales_occupation, palette = palette)
```

```python
#Product Category
```

```python
df.columns
```

```python
sns.set({'figure.figsize': (25, 5)})
palette = sns.color_palette("inferno", n_colors=18)
ax = sns.countplot(data = df, x = 'Product_Category', palette = palette)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#from above graph we can see that most of the sold products are from food colthing & apperial & Electronic & Gadages
```

```python
sales_productid = df.groupby(['Product_ID'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending=False).head(10)
sns.set({'figure.figsize': (15, 5)})
palette = sns.color_palette("inferno", n_colors=10)
sns.barplot(x = 'Product_ID', y = 'Orders', data = sales_productid, palette = palette)
```
